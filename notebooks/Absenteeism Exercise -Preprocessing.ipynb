# Transformation de type de la colonne

```python
type(df_reason_mod['Date'])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format ='%d/%m/%Y')
```

 # Extraire le mois d’une date

```python
df_reason_mod['Date'][0].month
```

```python
# Creation d'une liste vide pour pouvoir la remplir par la suite
list_months = []
list_months
```

```python
df_reason_mod.shape
```

```python
# À l'aide d'une boucle, on va pouvoir remplir la liste
for i in range (df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod
```

### Extraire le jour d'une date :

```python
df_reason_mod['Date'][699].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'],axis=1)
```

```python
df_reason_mod
```

```python
df_reason_mod.columns.values
```

```python
df_reason_mod_reorder = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[df_reason_mod_reorder]
```
