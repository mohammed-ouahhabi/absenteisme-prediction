```python
# Étape 1 : Importer les bibliothèques
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Étape 2 : Charger les données prédictives
df = pd.read_csv('Absenteeism_predictions.csv')
```

```python
# Étape 3 : Afficher les premières lignes
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))  # 1. Définir la taille du graphe
plt.scatter(df['Age'], df['Probability'], alpha=0.6, color='blue')  # 2. Créer le nuage de points
plt.xlabel('Age')  # 3. Nom de l’axe X
plt.ylabel('Probability of Being Absent')  # 4. Nom de l’axe Y
plt.title('Age vs Probability of Absence')  # 5. Titre du graphe
plt.grid(True)  # 6. Ajouter une grille
plt.show()  # 7. Afficher le graphe
```

```python
print(df.columns.tolist())
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. On récupère les données
ages = df['Age']
probs = df['Probability']

# 2. Calcul de la régression linéaire
slope, intercept = np.polyfit(ages, probs, 1)
line = slope * ages + intercept

# 3. Affichage du graphique
plt.figure(figsize=(10, 6))
plt.scatter(ages, probs, alpha=0.6, color='blue', label='Observations')
plt.plot(ages, line, color='red', linewidth=2, label='Ligne de tendance')

plt.xlabel('Age')
plt.ylabel('Probability of Being Absent')
plt.title('Age vs Probability of Absence with Trend Line')
plt.legend()
plt.show()
```
